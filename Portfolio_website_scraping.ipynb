```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('display.expand_frame_repr',False)
pd.set_option('max_colwidth',-1)
import pandas as pd
from bs4 import BeautifulSoup
import openpyxl
import requests
```

```python
url = 'https://danishsayyed.in/tableau'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
containers = soup.find_all('div', class_='x-el x-el-div page-inner c1-1 c1-2 c1-i c1-j c1-m c1-n c1-o c1-p')
data = {"titles": [], "Description": [], 'Image_url': [], 'Dashboard_url':[]}

for container in containers:
    title_data = container.find_all('h4')
    individual_title = [d.text.strip() for d in title_data] if title_data else []
        
    desc_data = container.find_all('div', class_='x-el c1-1 c1-2 c1-21 c1-22 c1-7s c1-2f c1-2g c1-7t c1-7u c1-7v c1-7w c1-7x c1-7y c1-7z c1-80 c1-81 c1-82 c1-83 c1-84 c1-85 c1-86 c1-87 c1-88 c1-89 c1-8a c1-8b c1-8c c1-8d c1-8e c1-8f c1-8g c1-8h c1-8i c1-8j c1-8k c1-6o c1-8l c1-i c1-5u c1-j c1-43 c1-m c1-n c1-o c1-p x-rt')
    individual_desc = [d.text.strip() for d in desc_data] if desc_data else []
    
    img_tags = container.find_all('img')
    Img = [img_tag.get('src', '') for img_tag in img_tags] if img_tags else []
    
    a_tag = container.find_all('a', class_ = 'x-el x-el-a c1-3r c1-1r c1-24 c1-8r c1-8s c1-1k c1-1o c1-1p c1-20 c1-22 c1-21 c1-u c1-1a c1-4 c1-3n c1-8t c1-8u c1-8v c1-8w c1-79 c1-8x c1-7n c1-8y c1-8z c1-90 c1-91 c1-92 c1-i c1-2l c1-93 c1-42 c1-94 c1-95 c1-96 c1-97 c1-44 c1-45 c1-46 c1-47')
    link = [l.get('href', '') for l in a_tag] if a_tag else []
    
    max_length = max(len(individual_title),len(individual_desc))
    data['titles'].extend(individual_title + [None] * (max_length - len(individual_title)))
    data['Description'].extend(individual_desc + [None] * (max_length - len(individual_desc))) 
    data['Image_url'].extend( Img + [None] * (max_length - len(Img)))
    data['Dashboard_url'].extend( link + [None] * (max_length - len(link)))
df = pd.DataFrame(data)
```
